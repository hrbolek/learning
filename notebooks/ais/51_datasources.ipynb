## CSV

```python
groups = [
    {"id": "3f1e81b6-561a-11ee-9e47-0242ac140004", "name": "G1"},
    {"id": "3f1e81f2-561a-11ee-9e47-0242ac140004", "name": "G2"}
]
```

```python
keys = groups[0].keys()
for key in keys:
    print(key)
```

```python
csvheader = ", ".join(groups[0].keys())
print(csvheader)
```

```python
csvlines = []
for item in groups:
    dataline = (item[key] for key in keys)
    csvline = ", ".join(dataline)
    csvlines.append(csvline)
    
print(csvlines)
csvfile = "\n".join([csvheader] + csvlines)
with open("./data.csv", "w") as jf:
    jf.write(csvfile)
```

```python
with open("./data.csv", "r") as jf:
    lines = jf.readlines()
lines
```

## HTML

```python
import aiohttp

async with aiohttp.ClientSession() as session:
    async with session.get('https://unob.cz') as resp:
        print(resp.status)
        textresponse = await resp.text()
        print(textresponse[:1000])
```

```python
async with aiohttp.ClientSession() as session:
    async with session.get('https://unob.cz') as resp:
        async for line in resp.content:
            print(line)
            print(line.decode("utf-8"))
            break
```

```python
async with aiohttp.ClientSession() as session:
    async with session.get('https://unob.cz') as resp:
        async for line in resp.content.iter_chunked(1024):
            print(line)
            print(30*"=")
            print(line.decode("utf-8"))
            break
```

# Filter, Map, Reduce

## Lambda

```python
addOne = lambda n: n + 1
data = [0, 1, 2, 3]

addOne(data[0])
```

## Filter

```python
nonZero = lambda n: n != 0
data = [0, 1, 2, 3]
list(filter(nonZero, data))
```

## Map

```python
asDict = lambda n: {'value': n}
data = [0, 1, 2, 3]
list(map(asDict, data))
```

## Reduce

```python
from functools import reduce

data = [0, 1, 2, 3]
sumIt = lambda acc, val: acc + val
reduce(sumIt, data)
```

```python
mulIt = lambda res, val: res * val
data = [0, 1, 2, 3]
reduce(mulIt, data)
```

```python
mulIt = lambda res, val: res * val
nonZero = lambda n: n != 0
data = [0, 1, 2, 3]

reduce(mulIt, filter(nonZero, data))
```

```python
mulIt = lambda res, val: res * val
nonZero = lambda n: n != 0
data = [0, 1, 2, 3]

reduce(mulIt, filter(nonZero, data), 0.5)
```
