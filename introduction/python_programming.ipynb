# Základy programování v Pythonu

## Jednoduché datové struktury

```
# cele cislo
a = 4
print(a)
```

```
# desetinne cislo
b = 1.4
print(b)
```

```
# textovy retezec
cstr = "Ahoj"
print(cstr)
dstr = 'Nazdar'
print(dstr)
```

```
# bool
cbool = True
dbool = False
print(cbool)
print(dbool)
```

## Složené datové struktury

```
# list / vektor
e = [1, 2, 3]
print(e)
f = ['Nazdar', "svete"]
print(f)
```

```
# dictionary
g = {'name': 'Josef', 'age': 5}
print(g)
print(g['name'])
```

## Řídící struktury

### Podmínky

```
# if
b = 1.4
if b < 2:
  print(b, ' < 2')
else:
  print(b, '>= 2')
```

```
# "multi" if
b = 1.4
if b < 1:
  print(b, ' < 1')
elif b < 2:
  print(b, ' < 2')
else:
  print(b, '>= 2')  
```

### Cykly

```
# for
data = [1, 2, 3, 4]
for item in data:
  print(item)
```

```
# for s indexem, porovnej s predchazejicim
for index, item in enumerate(data):
  print(index, item)
```

```
# https://docs.python.org/3/reference/compound_stmts.html?highlight=while#the-while-statement
# while
item = 0
while item < 3:
  item = item + 1
  print(item)
```

```
# while with break
item = 0
while True:
  item = item + 1
  print(item)
  if item >= 3:
    break
  
```
