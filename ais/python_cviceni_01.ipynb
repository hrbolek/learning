```python
def bind(func, firstParamValue):
    def G(#neco, #neco
          ):
        #neco
    return G

def add(x, y):
    return x + y

inc = bind(add, 1)
print(inc(5))
```

### Řešení

```python
#@title
def bind(func, firstParamValue):
    def G(*args, **kwargs):
        return func(firstParamValue, *args, **kwargs)
    return G

def add(x, y):
    return x + y

inc = bind(add, 1)
print(inc(5))
```
